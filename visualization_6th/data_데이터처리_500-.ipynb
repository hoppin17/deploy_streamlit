```python
df1.drop(columns = 'Unnamed: 0', inplace = True)
df1
```

```python
df1['actor_main'][0]
```

```python
import re

temp_lst = []

for lst in df1['actor_main']:

    # |을 이용하여 데이터 스플릿
    actor_names = lst.split('|')
    
    result = []
    # 각 배우의 이름만 추출하여 리스트에 저장
    for name in actor_names:
        result.append(re.sub(r'\[([^]]+)\]|\(([^)]+)\)|\s+', '', name))
    #print(result)    
    
    temp_lst.append(result)



df1['actor_main_name'] = temp_lst

# # 이름만 들어가진 lst로 채웠다.
df1['actor_main_name']
```

```python
# 임의로 한 위치 확인
df1.loc[197]
```

```python
df1[df1['actor_sub'] == np.nan]
```

```python
# nan으로 처리가된 부분이 있으므로 이부분도 신경써야한다.
df1[df1['actor_sub'].isnull() == True]
```

```python
# 조연도 변경

temp_lst = []

for lst in df1['actor_sub']:
    if not pd.isna(lst): # 조연은 없는 경우가 있으니 변경
        # |을 이용하여 데이터 스플릿
        actor_names = lst.split('|')
        
        result = []
        # 각 배우의 이름만 추출하여 리스트에 저장
        for name in actor_names:
            result.append(re.sub(r'\[([^]]+)\]|\(([^)]+)\)|\s+', '', name))
        #print(result)    
        
        temp_lst.append(result)
        
    else:
        temp_lst.append('null')



df1['actor_sub_name'] = temp_lst

# # 이름만 들어가진 lst로 채웠다.
df1['actor_sub_name']
```

```python
df1.head(40)
# 아직까지는 큰 문제가 없어보인다.
```

```python
df1[df1['actor_cameo'].isnull() == True]
```

```python
# 조연과 카메오도 같이 변경

temp_lst = []

for lst in df1['actor_cameo']:
    if type(lst) != float and '단역' not in lst: # 이경우만 특별출연 or 우정출연
        # |을 이용하여 데이터 스플릿
        actor_names = lst.split('|')
        
        result = []
        # 각 배우의 이름만 추출하여 리스트에 저장
        for name in actor_names:
            result.append(re.sub(r'\[([^]]+)\]|\(([^)]+)\)|\s+', '', name))
        #print(result)    
        
        temp_lst.append(result)
        
    else: # 아닌 경우는 null로 채움
        temp_lst.append('null')



df1['actor_cameo_name'] = temp_lst

# # 이름만 들어가진 lst로 채웠다.
df1['actor_cameo_name']
```

```python
temp_list = []
for num in range(len(df1['maker1'])):
    if type(df1['maker1'][num]) != float and '등급' not in df1['maker1'][num]:
        company = df1['maker1'][num].split('   ')
        temp_list.append(company)
    # 두 경우만 위치가 안맞아서 따로 넣어줬다. 이 부분만 코드를 다시짜는 게 훨신 번거로우니...
    elif num == 234:
        temp_list.append(['(주)오돌또기 Odoltogi', '(주)명필름 Myung Films'])
    elif num == 496:
        temp_list.append(['(주)드림써치C&C Dream Search Contents and Company co.Ltd','한국교육방송공사 EBS EBS   올리브스튜디오 Olive Studio'])
    else:
        if type(df1['maker1'][num]) != float and '등급' not in df1['maker2'][num]:
            company = df1['maker2'][num].split('   ')
            temp_list.append(company)
        else: # 지금 값이 없는 것을 제거한다
            temp_list.append('null')
            

print(len(temp_list))        
```
