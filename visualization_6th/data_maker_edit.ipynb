```python
exclude_list = ['51k', 'MBC씨앤아이', '나의PS파트너 문화산업전문회사']
```

```python
import re
```

```python
maker_temp = []
for x in range(499):
    # 각각의 리스트를 불러옴
    try1 = eval(df2['maker'][x])
    # 각각을 저장할 임시 리스트
    temp = []
    for y in try1:
        # 위와 같은 경우는 수정X
        if y in exclude_list:
            temp.append(y)
        else:
            cleaned_text = re.sub(r'[a-zA-Z]', '', y)
            cleaned_text = re.sub(r'[!@#$%^&*()_+={}\[\]:;"\'<>,.?\\/]', '', cleaned_text)
            cleaned_text = cleaned_text.strip()
            temp.append(cleaned_text)
    maker_temp.append(temp)
maker_temp    
```

```python
df2['maker'] = maker_temp
```

```python
df2['gener'] = df1['only_gener']
```
